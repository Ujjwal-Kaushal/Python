```python
import re # Make sure you have this import
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# Initialize stemmer
ps = PorterStemmer()

# Convert stopwords to a set for faster lookup (more efficient than a list)
all_stopwords = set(stopwords.words('english'))

corpus = []
for i in range(0, len(messages)):
    # Step 1: Get the current raw message from your DataFrame
    current_message = messages['message'][i]

    # Step 2: Clean the text - Remove non-alphabetic characters and prepare for tokenization
    # Use '[^a-zA-Z]' to replace anything that is NOT an English letter (both lowercase and uppercase)
    # with a single space. This handles punctuation, numbers, etc.
    review = re.sub('[^a-zA-Z]', ' ', current_message)
    
    # Convert the entire string to lowercase AFTER cleaning for consistency.
    # This ensures "Hello" becomes "hello", not " ello"
    review = review.lower()

    # Step 3: Tokenize - Split the string into a list of individual words
    # .split() without arguments splits by any whitespace and handles multiple spaces,
    # ensuring you get clean words like ['hello', 'world'] from "hello  world"
    words = review.split()

    # Step 4: Remove stopwords and Step 5: Perform stemming
    processed_words = []
    for word in words:
        # Check if the word is NOT in our set of stop words
        if word not in all_stopwords:
            # Stem the entire word (not just a character)
            stemmed_word = ps.stem(word)
            processed_words.append(stemmed_word)

    # Step 6: Rejoin - Join the processed words back into a single string, separated by a single space
    processed_review = ' '.join(processed_words)
    
    corpus.append(processed_review)

# After running this code, your 'corpus' list will contain properly processed reviews.
# You can then print elements from 'corpus' to see the corrected output:

# Example: Print the first 5 processed messages
for j in range(0, len(corpus)):
    print(f"Processed message {j+1}: {corpus[j]}\n")

# Or print a specific one:
# print(corpus[0])
```

```python
# Create bag of words model 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 900, binary = True, ngram_range = (1,1))
X = cv.fit_transform(corpus).toarray()
```
